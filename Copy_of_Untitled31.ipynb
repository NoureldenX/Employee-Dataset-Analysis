```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
df = pd.read_csv("/content/Employee.csv")
```

```python
df.fillna(0, inplace=True)
```

```python
x = df.drop(columns=['ExperienceInCurrentDomain'])  # Features
y = df['ExperienceInCurrentDomain']  # Target
```

```python
print("First 5 rows:\n", df.head())
print("\nShape:", df.shape)
print("\nColumns:", df.columns)
print("\nData Types:\n", df.dtypes)
print("\nMissing Values:\n", df.isnull().sum())
print("\nSummary statistics:\n", df.describe())
```

```python
print("\nEducation:\n", df["JoiningYear"].value_counts())
print("\nPaymentTier:\n", df["Age"].value_counts())
print("\nGender:\n", df["EverBenched"].value_counts())
print("\nExperienceInCurrentDomain:\n", df["Education"].value_counts())
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='PaymentTier', hue='JoiningYear', palette='viridis')
plt.title('Employee Count per Department')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.hist(df['PaymentTier'], bins=30, color='teal', edgecolor='black')
plt.title('Distribution of Monthly Income')
plt.xlabel('Monthly Income')
plt.ylabel('Number of Employees')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='JoiningYear', y='Age', palette='coolwarm')
plt.title('Monthly Income by Job Role')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
fig = px.pie(df, names='JoiningYear', title='PaymentTier', color_discrete_sequence=px.colors.sequential.RdBu)
fig.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='JoiningYear', y='PaymentTier', palette='pastel')
plt.title('Job Satisfaction by Attrition')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Age', palette='Set2')
plt.title('Paymentier')
plt.tight_layout()
plt.show()
```

```python
fig = px.scatter(
    df,
    x="Age",
    y="PaymentTier",
    color="LeaveOrNot",
    title="Age vs Payment Tier by Attrition",
    hover_data=["Education", "City"]
)
fig.show()
```
